```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import datetime
```

```python
FILENAME = 'cab_rides.csv'
df = pd.read_csv(FILENAME)
```

```python
#Changing time Stamp to date and time
df['time_stamp'] = pd.to_datetime(df['time_stamp'], unit='ms').dt.time
```

```python
df
```

```python
#sorting the uber and Lyft rides
df_1 = df.loc[df['cab_type']=='Uber']
df_2 = df.loc[df['cab_type']=="Lyft"]
```

```python
# Sorting distance of both the datasets
#Uber
distances_Uber = (df_1.distance)
distances_Uber_unique = sorted(df_1.distance.unique())
#Lyft
distances_Lyft = (df_2.distance)
distances_Lyft_unique = sorted(df_2.distance.unique())

#Prices
#Uber
price_Uber = (df_1.price)
#Lyft
price_Lyft = (df_2.price)

#time
#Uber
time_Uber = (df_1.time_stamp)
#Lyft
time_Lyft = (df_2.time_stamp)








```

```python
### Uber Dataframe sorted by time
Dict_Uber = {'Distance':distances_Uber,'Price':price_Uber,'Time':time_Uber}
df_Uber1 = pd.DataFrame(Dict_Uber)
df_Uber = df_Uber1.sort_values('Time').dropna()
df_Uber

```

```python
###LYFT DataFrame sorted by time
Dict_Lyft = {'Distance':distances_Lyft,'Price':price_Lyft,'Time':time_Lyft}
df_Lyft1 = pd.DataFrame(Dict_Lyft)
df_Lyft = df_Lyft1.sort_values('Time').dropna()
df_Lyft
```

```python
#Uber distance and prices and time list
Uber_distance = df_Uber.Distance.tolist()
Uber_price = df_Uber.Price.tolist()
u_time = df_Uber.Time.tolist()





```

```python
#Lyft distance and prices and time list
Lyft_distance = df_Lyft.Distance.tolist()
Lyft_price = df_Lyft.Price.tolist()
l_time = df_Lyft.Time.tolist()

```

```python
#Converting Uber time to string and dividing the time periods into 4 parts
Uber_time =[]
for i in range(len(u_time)):
    v = u_time[i].strftime("%H:%M:%S")
    Uber_time.append(v)

Uber_00to06 = []
Uber_06to12 = []
Uber_12to18 = []
Uber_18to00 = []
Uber_00to06_price = []
Uber_06to12_price = []
Uber_12to18_price = []
Uber_18to00_price = []
Uber_00to06_distance = []
Uber_06to12_distance = []
Uber_12to18_distance = []
Uber_18to00_distance = []


for i in range(len(Uber_time)):
    if('00:00:00'<= Uber_time[i]<'06:00:00'):
        Uber_00to06.append(Uber_time[i])
        Uber_00to06_price.append(Uber_price[i])
        Uber_00to06_distance.append(Uber_distance[i])
    if('06:00:00'<= Uber_time[i]<'12:00:00'):
        Uber_06to12.append(Uber_time[i])
        Uber_06to12_price.append(Uber_price[i])
        Uber_06to12_distance.append(Uber_distance[i])
    if('12:00:00'<= Uber_time[i]<'18:00:00'):
        Uber_12to18.append(Uber_time[i])
        Uber_12to18_price.append(Uber_price[i])
        Uber_12to18_distance.append(Uber_distance[i])
    if('18:00:00'<= Uber_time[i]<='23:59:59'):
        Uber_18to00.append(Uber_time[i])
        Uber_18to00_price.append(Uber_price[i])
        Uber_18to00_distance.append(Uber_distance[i])

#Converting Lyft time to string and dividing the time periods into 4 parts

Lyft_time =[]
for i in range(len(l_time)):
    c = l_time[i].strftime("%H:%M:%S")
    Lyft_time.append(c)

Lyft_00to06 = []
Lyft_06to12 = []
Lyft_12to18 = []
Lyft_18to00 = []
Lyft_00to06_price = []
Lyft_06to12_price = []
Lyft_12to18_price = []
Lyft_18to00_price = []
Lyft_00to06_distance = []
Lyft_06to12_distance = []
Lyft_12to18_distance = []
Lyft_18to00_distance = []

for i in range(len(Lyft_time)):
    if('00:00:00'<= Lyft_time[i]<'06:00:00'):
        Lyft_00to06.append(Lyft_time[i])
        Lyft_00to06_price.append(Lyft_price[i])
        Lyft_00to06_distance.append(Lyft_distance[i])
    if('06:00:00'<= Lyft_time[i]<'12:00:00'):
        Lyft_06to12.append(Lyft_time[i])
        Lyft_06to12_price.append(Lyft_price[i])
        Lyft_06to12_distance.append(Lyft_distance[i])
    if('12:00:00'<= Lyft_time[i]<'18:00:00'):
        Lyft_12to18.append(Lyft_time[i])
        Lyft_12to18_price.append(Lyft_price[i])
        Lyft_12to18_distance.append(Lyft_distance[i])
    if('18:00:00'<= Lyft_time[i]<='23:59:59'):
        Lyft_18to00.append(Lyft_time[i])
        Lyft_18to00_price.append(Lyft_price[i])
        Lyft_18to00_distance.append(Lyft_distance[i])
    

   



    

        
```
